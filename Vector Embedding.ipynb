```python
import numpy as np

# words
word1 = 'man'
word2 = 'woman'

word3 = 'semiconductor'
word4 = 'earthworm'

word5 = 'nephew'
word6 = 'niece'

# get vectors from the model
vec1 = model[word1]
vec2 = model[word2]

vec3 = model[word3]
vec4 = model[word4]

vec5 = model[word5]
vec6 = model[word6]

# difference vectors
diff1 = vec1 - vec2
diff2 = vec3 - vec4
diff3 = vec5 - vec6

# magnitudes (L2 norm)
mag1 = np.linalg.norm(diff1)
mag2 = np.linalg.norm(diff2)
mag3 = np.linalg.norm(diff3)

# print results
print(f"|| {word1} - {word2} || = {mag1:.4f}")
print(f"|| {word3} - {word4} || = {mag2:.4f}")
print(f"|| {word5} - {word6} || = {mag3:.4f}")

```
